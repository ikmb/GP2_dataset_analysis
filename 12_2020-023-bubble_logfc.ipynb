# Preparing the merged_df


```python
res_prim_filt = res_prim_stool[(res_prim_stool['passed_ss_GroupPSC'] == True) & (res_prim_stool['q_GroupPSC'] < 0.1)]
```

```python
asv_da = asv_da[['taxon', 'lfc_GroupPSC', 'q_GroupPSC']]
```

```python
asv_da.set_index('taxon', inplace=True)
```

```python
asv_da
```

```python
merged_df = pd.concat([asv_da, PSC_tax_stool[['Genus', 'Species']]], axis=1)
```

```python
merged_df
```

```python
merged_df['Genus'] = merged_df['Genus'].str.replace('/', '.')
```

```python
subset_genus = merged_df[merged_df['Genus'].isin(res_prim_filt['taxon'])]
```

```python
subset_genus
```

```python
subset_genus[subset_genus['Genus'] == 'G__Prevotella']
```

```python
subset_genus.dropna(subset=['lfc_GroupPSC'], inplace=True)
```

```python
subset_genus
```

```python
# Get unique Genus values
unique_genus = subset_genus['Genus'].unique()

# Create a dictionary to map each unique Genus value to a color from tab10 palette
genus_color_map = {genus: plt.cm.tab10(i) for i, genus in enumerate(unique_genus)}

# Assign ranks to q_GroupPSC values
subset_genus['q_GroupPSC_rank'] = 25 - subset_genus['q_GroupPSC'].rank()

# Initialize lists to store data for the bubble plot
x = []
y = []
color = []
size = []

# Iterate over each row in subset_genus
for index, row in subset_genus.iterrows():
    # Append logFC values to x
    x.append(row['lfc_GroupPSC'])
    
    # Append index values to y
    y.append(index)
    
    # Get color for the Genus from the dictionary
    color.append(genus_color_map[row['Genus']])
    
    # Calculate bubble size based on rank of q_GroupPSC values
    size.append(row['q_GroupPSC_rank'] * 100)  # Adjust multiplier as needed



# Create the bubble plot
plt.figure(figsize=(10, 8))
# Add background grid lines
plt.grid(True, linestyle='--', alpha=0.5)
scatter = plt.scatter(x=x, y=y, c=color, s=size, alpha=0.8)

# Add labels for each point
for i, txt in enumerate(subset_genus['Species']):
    if not pd.isna(txt):
        plt.annotate(txt, (x[i], y[i]), fontsize=8, ha='center', va='center')
    

# Add labels and title
plt.xlabel('logFC')
plt.ylabel('ASV')

# Show legend
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=label) for label, color in genus_color_map.items()]
plt.legend(handles=handles, title='Genus', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.savefig('./results/bubble_diff_abund_stool.png', dpi=300, bbox_inches='tight')
# Show plot
plt.show()
```
